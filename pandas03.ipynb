># **15- FILTERING (Sorting) a DataSet**

```python
df_DataForFilter.embark_town.unique()
```

```python
df_DataForFilter[(df_DataForFilter.embark_town == "Southampton")]
```

>#### If you want to filter multiple values in a column, use (.isin([...])) method
 

```python
df_DataForFilter[df_DataForFilter.embark_town.isin(['Southampton','Cherbourg'])]
```

># **16- use AND , OR , NOT (& , | , !=) FOR FILTERING**

```python
df_DataForFilter[(df_DataForFilter.embark_town=='Southampton') & # Filter rows where embark_town is Southampton AND alive is yes
                 (df_DataForFilter.alive=='yes')]
```

```python
df_DataForFilter[(df_DataForFilter.embark_town=='Southampton') &  
                 (df_DataForFilter.age>30)]   # Show rows where embark_town is Southampton AND age is greater than 30
```

```python
df_DataForFilter[df_DataForFilter.embark_town != 'Southampton']
# Same: Show rows where embark_town is NOT Southampton
```

```python
df_DataForFilter[(df_DataForFilter.embark_town == 'Southampton') | 
                 (df_DataForFilter.embark_town == 'Cherbourg')]
# Show rows where embark_town is Southampton OR Cherbourg
```

># **22- Reshape multiindex series**


```python
df_DataForFilter.describe()
```

```python
df_DataForFilter.survived.mean()
```

```python
# Group the data by 'who' and 'age' and calculate the average survival rate for each group
df_DataForFilter.groupby(['who','age']).survived.mean()
```

```python
# Group by 'who' and 'pclass', calculate average survival rate,and use unstack to look clean 
df_DataForFilter.groupby(['who','pclass']).survived.mean().unstack()
```

```python
# then reshape the result so 'who' becomes columns and 'age' becomes rows female and child high survied rate then man 
df_DataForFilter.groupby(['who','pclass']).survived.mean().unstack().T
```

># **26- Reshapeing a dataframe**

```python
# Convert from wide to long format:
# Keep 'Name' as identifier
# Combine 'Math' and 'Science' columns into one 'Subject' column
# Move their values into a new column called 'Score'
df_reshapeData.melt(id_vars='Name', var_name='Subject', value_name='Score')
```
